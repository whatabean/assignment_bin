```python
import pyBigWig
import matplotlib.pyplot as plt

bw = pyBigWig.open("./fastq_data/sample.bw")
chrom = "NC_000002.12"

start, end = 25000000, 40000000
values = bw.values(chrom, start, end)
```

```python
plt.figure(figsize=(10, 3))
plt.plot(range(start, end), values)
plt.title(f"ChIP-seq signal on {chrom}:{start}-{end}")
plt.xlabel("Genomic position")
plt.ylabel("Coverage")
plt.show()
```
